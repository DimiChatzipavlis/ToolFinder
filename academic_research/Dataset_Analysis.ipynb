# Synthetic Dataset Exploration: MCP Neural Routing

This notebook validates the two training corpora, `mcp_routing_dataset.csv` and `mcp_routing_dataset_v2.csv`, generated from live GitHub MCP server schemas. It documents the anchor-generation policy used during dataset construction, where natural language queries were synthesized using a 30/40/30 variance rule: 30% explicit commands, 40% implicit questions, and 30% conversational context.

```python
import pandas as pd
import json
```

## Dataset V1: Core GitHub Tools

```python
df_v1 = pd.read_csv('mcp_routing_dataset.csv')
print(f'Total rows: {len(df_v1)}')

tool_names_v1 = []
for schema in df_v1['positive_schema']:
    parsed = json.loads(schema)
    tool_names_v1.append(parsed['name'])

unique_tools_v1 = sorted(set(tool_names_v1))
print('Unique V1 tools:')
for index, tool_name in enumerate(unique_tools_v1, start=1):
    print(f'{index}. {tool_name}')

print('\nFirst 5 anchors:')
for index, anchor in enumerate(df_v1['anchor'].head(5), start=1):
    print(f'{index}. {anchor}')
```

## Dataset V2: Extended MCP Capabilities

```python
df_v2 = pd.read_csv('mcp_routing_dataset_v2.csv')
print(f'Total rows: {len(df_v2)}')

tool_names_v2 = []
for schema in df_v2['positive_schema']:
    parsed = json.loads(schema)
    tool_names_v2.append(parsed['name'])

unique_tools_v2 = sorted(set(tool_names_v2))
print('Unique V2 tools:')
for index, tool_name in enumerate(unique_tools_v2, start=1):
    print(f'{index}. {tool_name}')
```

Combined, these datasets provide 1,500 highly varied natural language queries mapped perfectly to 30 distinct Model Context Protocol JSON schemas, forming a robust foundation for Bi-Encoder contrastive learning.
